```python
#Using ScraperAPI Website to get an API key
import requests # Used to make HTTP request to the twitter website
import pandas as pd
```

```python
# create an empty list to append our data 
twitter_data = []
```

```python
# A set that will contain the parameters that will dectate how our search works
payload = {
    'api_key':'179d903c5ff3b887fa1ea4b0beba4c43',
    'query':'Amazon products',
    'date_range_start':'2022-11-14',
    'date_range_end':'2023-11-14',
    'num':'300'
}
response = requests.get(
    'https://api.scraperapi.com/structured/twitter/search',params=payload# passing the link to the scraper API
)
data = response.json()
```

```python
#Iterating through our data and put it in the dataframe
all_tweets = data['organic_results']# this is a key from our data
for tweet in all_tweets:
    twitter_data.append(tweet)
```

```python
df = pd.DataFrame(twitter_data)
df.to_csv('tweetsAnalysis11.csv')
print("file exported")
```
